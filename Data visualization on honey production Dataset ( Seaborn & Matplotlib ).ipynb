## 3. Display the percentage distribution of the data in each year using the pie chart.

```python
df.groupby('year').count()[['state']].plot(y='state',kind='pie',figsize=(4,4),autopct='%1.1f%%',legend=False)
```

## 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.displot(x=df['priceperlb'],kind='hist',bins=30,kde=True)
plt.show()                                                           


##Findings:

## using this histogram plot for ['priceperlb']and the distribution of the graph is peakin somepoint.
## where the [pricerperlb] value in between 1.3 to 1.5 and count value in between 50 to 60.
## so the graph is positively Skewed..
```

## 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, andwrite your findings.


```python
plt.scatter(x=df['numcol'],y=df['prodvalue'])
plt.title('numcol v/s prodvalue')
plt.show()    



## findings:

## The relationship between 'numcol'vs'prodvalue'is directly related with then representationof graph.
## The of 'numcol' and 'prodvalue' is merged in somepoint.
## The graph values is also gradually increasing.
```

## 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable'prodvalue' through boxplot, and write your findings.

```python
sns.catplot(data=df,x='year',y='prodvalue',kind='box',height=5,aspect=1.2)
plt.show()


## Findings:
# From this categorical plotting'numcol', 'totalprod' and 'prodvalue' are positively skewed and 
#'yieldpercol' seems to be normally distributed and the 
#'totalprod' vs 'numcol' and 'totalprod' vs 'prodvalue' seems to be directly(positively) realted.
```

## 7. Visualize and understand the relationship between the multiple pairs of variables throughout different yearsusing pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df[['numcol','yieldpercol','totalprod','prodvalue','year']],corner=True)


## Findings:

## 'numcol','totalprod','prodvalue' are positively skewed.
## 'yieldpercol' are seems to be normally distributed graph
## So the 'numcol','totalprod','prodvalue' are positively skewed with each other.
```

## 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')
